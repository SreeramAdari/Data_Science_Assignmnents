```python
import numpy as np
from scipy import stats

mu_x = 600
sigma_x = 25
n = 25
observed_mean_cost = 3050

predicted_mean_cost = 1000 + 5 * mu_x

standard_error = (5*sigma_x)/np.sqrt(n)

t_statistic = (observed_mean_cost - predicted_mean_cost) / standard_error

df = n - 1

alpha = 0.05
critical_value = stats.t.ppf(1 - alpha/2, df)

p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))


print(f"Observed Mean Cost: {observed_mean_cost}")
print(f"Predicted Mean Cost: {predicted_mean_cost}")
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
print(f"Standard Error: {standard_error}")
print(f"Critical Value: {critical_value}")


if p_value < alpha:
    print(f"Reject the null hypothesis. The observed costs are significantly different from the predicted costs at significance level {alpha}.")
    if observed_mean_cost > predicted_mean_cost:
      print("There is strong evidence to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests.")
    else:
      print("There is strong evidence to support that the weekly operating costs are lower than the model suggests.")
elif abs(t_statistic) > critical_value:
    print(f"Reject the null hypothesis based on critical value. The observed costs are significantly different from the predicted costs.")
    if observed_mean_cost > predicted_mean_cost:
      print("There is strong evidence to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests.")
    else:
      print("There is strong evidence to support that the weekly operating costs are lower than the model suggests.")
else:
    print(f"Fail to reject the null hypothesis. There is not enough evidence to suggest the observed costs are significantly different from the predicted costs at significance level {alpha}.")
```
