```python
def is_prime(number):
  if number <= 1:
    return False
  for i in range(2, int(number**0.5) + 1):
    if number % i == 0:
      return False
  return True
num = int(input("Enter a number: "))
if is_prime(num):
  print(f"{num} is a prime number")
else:
  print(f"{num} is not a prime number")
```

```python
import random

def generate_problem():
  num1 = random.randint(1, 10)
  num2 = random.randint(1, 10)
  product = num1 * num2
  return num1, num2, product

num1, num2, product = generate_problem()

print(f"What is the product of {num1} and {num2}?")
user_answer = int(input("Your answer: "))


if user_answer == product:
  print("Correct!")
else:
  print(f"Incorrect. The correct answer is {product}")
```

```python
def print_even_squares(start, end):
  for num in range(start, end + 1):
    if num % 2 == 0:
      print(num**2)
print_even_squares(100, 200)
```

```python
import re
from collections import Counter

def count_words(input_text):
    text = re.sub(r'[^\w\s]', '', input_text).lower()
    words = text.split()
    word_counts = Counter(words)
    return word_counts

input_text = "This is a sample text. This text will be used to demonstrate the word counter."
word_counts = count_words(input_text)
for word, count in word_counts.items():
    print(f"'{word}': {count}")
```

```python
def is_palindrome(text):
  processed_text = ''.join(filter(str.isalnum, text)).lower()
  return processed_text == processed_text[::-1]

input_string = input("Enter a string: ")
if is_palindrome(input_string):
  print(f"'{input_string}' is a palindrome.")
else:
  print(f"'{input_string}' is not a palindrome.")
```
